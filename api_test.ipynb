```python
from skyscanner import Flights

flights_service = Flights('se893794935794863942245517499220')

result = flights_service.get_result(
    country='UK',
    currency='GBP',
    locale='en-GB',
    originplace='SIN-sky',
    destinationplace='KUL-sky',
    outbounddate='2016-07-28',
    inbounddate='2016-07-31',
    adults=1).parsed

print(result['Itineraries'][:10])
```

```python
from skyscanner import FlightsCache

flights_cache_service = FlightsCache('se893794935794863942245517499220')

result = flights_cache_service.get_cheapest_quotes(
    market='US',
    currency='USD',
    locale='en-US',
    originplace='DTW-sky',
    destinationplace='anywhere',
    outbounddate='2016-08',
    inbounddate='2016-08').parsed

print(result['Quotes'][:1])
```
